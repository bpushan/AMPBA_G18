```python
save_dir = '/content/drive/MyDrive/NLP/Dataset/'
os.makedirs(save_dir, exist_ok=True)
```

```python
## Function to download image from URL and save to Google Drive
def download_image(url, save_path):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            with open(save_path, 'wb') as f:
                f.write(response.content)
            return True
        else:
            print(f"Failed to download image from {url}. Status code: {response.status_code}")
            return False
    except Exception as e:
        print(f"Exception occurred while downloading image from {url}: {e}")
        return False
```

```python
## Iterate through each row in the DataFrame and download images
for i, (key, url, caption) in enumerate(zip(dataset['test']['image/key'], dataset['test']['image/url'], dataset['test']['IIW'])):
    image_name = f'{key}.jpg'  # Naming the image files based on the key
    image_path = os.path.join(save_dir, image_name)

    # Download the image and save it to Google Drive
    success = download_image(url, image_path)
    if success:
        print(f"Downloaded {url} to {image_path}")
    else:
        print(f"Failed to download {url}")
```
